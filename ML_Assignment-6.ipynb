```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import squareform, pdist
import warnings
warnings.filterwarnings("ignore")
```

```python
A1 = np.array([0.4005,0.2148,0.3457,0.2652,0.0789,0.4548])
A2 = np.array([0.5306,0.3854,0.3156,0.1875,0.4139,0.3022])
```

```python
point = ['P1','P2','P3','P4','P5','P6']
data = pd.DataFrame({'Point':point, 'x cordinate':A1, 'y cordinate':A2})
data = data.set_index('Point')
data
```

```python
dist = pd.DataFrame(squareform(np.round(pdist(data[['x cordinate', 'y cordinate']]),4), 'euclidean'), columns=data.index.values, index=data.index.values)
dist
```

```python
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram with Single inkage")  
dendogram = sch.dendrogram(sch.linkage(data[['x cordinate', 'y cordinate']], method='single'), labels=data.index)
```

```python
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram with Complete inkage")  
dendogram = sch.dendrogram(sch.linkage(data[['x cordinate', 'y cordinate']], method='complete'), labels=data.index)
```

```python
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram with Average inkage")  
dendogram = sch.dendrogram(sch.linkage(data[['x cordinate', 'y cordinate']], method='average'), labels=data.index)
```
